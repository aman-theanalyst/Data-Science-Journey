**Dictionary**

A dictionary in Python is a collection of key–value pairs.    
It’s used to store data in a structured way, where each value is accessed using its key, not an index.   
Dictionaries are written with curly brackets, and have keys and values.

**Characterstics:**
*   Mutable
*   Indexing has no meaning
*   keys can't be duplicated
*   keys can't be mutable items






# Create Dictionary

```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
print(thisdict)
```

```python
# 2D dictionary -> JSON

student_data = {
    'name':'nitish',
     'college':'bit',
     'sem':4,
     'subjects':{
         'dsa':50,
         'maths':67,
         'english':34
     }
}

student_data
```

```python
# Duplicate values will overwrite existing values

thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964,
  "year": 2020
}
print(thisdict)
```

# Accessing Item

```python
# 1. by referring to its key name, inside square brackets

thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
x = thisdict["model"]
print(x)
```

```python
# 2. using get()

thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
x = thisdict.get("year")
print(x)
```

```python
# keys() - return a list of all the keys in the dictionary.

thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
x = thisdict.keys()
print(x)
```

```python
# values() - return a list of all the values in the dictionary.

thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
x = thisdict.values()
print(x)
```

```python
# items() - return each item in a dictionary, as tuples in a list.

thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
x = thisdict.items()
print(x)
```

```python
# in keyword - to determine if a specified key is present in a dictionary

thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
if "model" in thisdict:
  print("Yes, 'model' is one of the keys in the thisdict dictionary")
```

# Adding key-value pair

```python
# using a new index key and assigning a value to it

thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict["color"] = "red"
print(thisdict)
```

```python
# update() - if the item does not exist, the item will be added.

thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict.update({"color": "red"})
```

# Change Item

```python
# by referring to its key name

thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict["year"] = 2018
print(thisdict)
```

```python
# update()

thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict.update({"year": 2020})

print(thisdict)
```

# Remove Item

```python
# pop() -removes the item with the specified key name

thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict.pop("model")
print(thisdict)
```

```python
# popitem() - removes the last inserted item

thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict.popitem()
print(thisdict)
```

```python
# del keyword - removes the item with the specified key name

thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
del thisdict["model"]
print(thisdict)


# The del keyword can also delete the dictionary completely
del thisdict
# print(thisdict) this will cause an error because "thisdict" no longer exists.
```

```python
# clear() - empties the dictionary:

thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict.clear()
print(thisdict)
```

# Loops In Dictionary

```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}

for x in thisdict:
  print(x)
```

# Copy In Dictionary

```python
# copy() -

thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
mydict = thisdict.copy()
print(mydict)
```

```python
# dict()

thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
mydict = dict(thisdict)
print(mydict)
```

# Nested Dictionary

```python
myfamily = {
  "child1" : {
    "name" : "Emil",
    "year" : 2004
  },
  "child2" : {
    "name" : "Tobias",
    "year" : 2007
  },
  "child3" : {
    "name" : "Linus",
    "year" : 2011
  }
}

print(myfamily)
```

# Dictionary Operation

*   Membership
*   Iteration

```python
# Membership - check if a key exists in the dictionary using the in and not in operators


d = {"name": "Aman", "age": 21, "city": "Delhi"}

print("name" in d)       # ✅ True (key exists)
print("email" not in d)  # ✅ True (key doesn't exist)
print("Delhi" in d.values())  # ✅ True (checking in values)
```

```python
# Iteration

d = {"name": "Aman", "age": 21, "city": "Delhi"}

# 1.Iterate through keys
for key in d:
    print(key)

# 2.Iterate through values
for value in d.values():
    print(value)

# 3.Iterate through key–value pairs
for key, value in d.items():
    print(key, "->", value)
```
